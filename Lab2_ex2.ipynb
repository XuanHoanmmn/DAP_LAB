```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
```

```python
wine = '/content/drive/MyDrive/DAP/wines.csv'
df = pd.read_csv(wine)
#xử lý khoảng trắng
df.columns = df.columns.str.replace(' ','_')
print(df.columns.tolist())
```

```python
#kiểm tra cấu trúc
print('\n')
df.info()
print('\n')
df.describe()
print('\n')
print(f"missing values:{df.isnull().sum()}")
```

```python
#Linear Regression
#dự đoán quality
#xây dựng ols
lm_model = smf.ols(formula='quality ~ alcohol + volatile_acidity + citric_acid + sulphates', data=df).fit()
print(lm_model.summary())
```

```python
#logistic regression,ex2.2
titanic = '/content/drive/MyDrive/DAP/titanic.csv'
df = pd.read_csv(titanic)
model_titanic = smf.logit(formula='survived ~ age + pclass + sex', data=df).fit()
print(model_titanic.summary())
```

```python
#Exercise 2.3 Multinomial
mn_model = smf.mnlogit(formula='quality ~ alcohol + volatile_acidity + citric_acid + sulphates', data=df).fit()
print(mn_model.summary())
```
