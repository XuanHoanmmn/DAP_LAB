```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
```

```python
wine = '/content/drive/MyDrive/DAP/wines.csv'
df = pd.read_csv(wine)
#xử lý khoảng trắng
df.columns = df.columns.str.replace(' ','_')
print(df.columns.tolist())
```

```python
#kiểm tra cấu trúc
print('\n')
df.info()
print('\n')
df.describe()
print('\n')
print(f"missing values:{df.isnull().sum()}")
```

```python
#Linear Regression
model = smf.ols(formula='quality~fixed_acidity + volatile_acidity + citric_acid + residual_sugar + chlorides + free_sulfur_dioxide + total_sulfur_dioxide + density + pH + sulphates + alcohol + type',data = df)
res = model.fit()
res.summary()
```

R2 và adj R2 thấp chỉ giải thích được sự biến thiên cả chất lượng rượu khoảng 30%.Khả năng dự đoán chính xác của model thấp

Xét P-Value:

citric_acid > 0.05 -> loại

chlorides > 0.5 -> loại

```python
#tạo lại model với formula đã bỏ biến citric_acid và chlorides với Pvalue > 0.05
model = smf.ols(formula='quality~fixed_acidity + volatile_acidity  + residual_sugar  + free_sulfur_dioxide + total_sulfur_dioxide + density + pH + sulphates + alcohol + type',data = df)
res = model.fit()
res.summary()
```

Sau khi bỏ 2 biến citric_acid và chlorides với Pvalue > 0.05 thì R2 và adj R2 vẫn không thay đổi.Và các biến đã có P_value hoàn toàn <0.05

#logistic regression,ex2.2

```python
titanic = '/content/drive/MyDrive/DAP/titanic.csv'
df = pd.read_csv(titanic)
```

```python
model = smf.glm(formula='survived~+C(pclass)+ sex + age + sibsp + parch + fare + C(embarked)',data=df)
res = model.fit()
res.summary()
```

Loại bỏ một số biến trước khi chạy model:

boat, body : gây overfitting

passenger_id, name, ticket, cabin,house.det : biến vô nghĩa

Sau khi chạy mô hình

Xét P_value:

parch > 0.05 -> loại

fare >0.05 -> loại

```python
model = smf.glm(formula='survived~+C(pclass)+ C(sex) + age + sibsp  + C(embarked)',data=df)
res = model.fit()
res.summary()
```

Các biến đã có P_Value < 0.05

sex là biến ảnh hưởng lớn nhất đến dự đoán khi coef cao và P_value = 0

#Exercise 2.3 Multinomial

```python
wine = '/content/drive/MyDrive/DAP/wines.csv'
df = pd.read_csv(wine)
#xử lý khoảng trắng
df.columns = df.columns.str.replace(' ','_')
print(df.columns.tolist())
```

```python
#Phân loại đa biến với quality có các mức đánh giá khác nhau như 4,5,6,..
import statsmodels.api as sm
model = sm.MNLogit.from_formula("quality ~ alcohol + pH + sulphates + density + chlorides + residual_sugar + total_sulfur_dioxide + free_sulfur_dioxide + citric_acid + volatile_acidity + fixed_acidity",data=df)
res = model.fit()
res.summary()
```

Xét P_value > 0.05:

citric_acid, total_sulfur_dioxide ,residual_suga -> loại


```python
#chạy lại mô hình sau khi bỏ các biến P_value > 0.05
model = sm.MNLogit.from_formula("quality ~ alcohol + volatile_acidity + chlorides + fixed_acidity + sulphates + pH + free_sulfur_dioxide",data = df)
res = model.fit()
res.summary()
```
