Lab_part2 Data preparation

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Exercise 1.1
#import medical_care data
titanic = '/content/drive/MyDrive/DAP/titanic.csv'
df = pd.read_csv(titanic)
print(f"shape:{df.shape}")
print('\n')
df.info()
print('\n')
df.head()
```

```python
df.head()
```

```python
#Exercise 1.2
# tạo dict để map
embarked_map = {
    'C': 'Cherbourg',
    'Q': 'Queenstown',
    'S': 'Southampton'
}
sex_map = {
    0: 'Male',
    1: 'Female'
}
survived_map = {
    0: 'No',
    1: 'Yes'
}
pclass_map = {
    1: '1st',
    2: '2nd',
    3: '3rd'
}
# mapping
df["embarked"] = df["embarked"].map(embarked_map)
df["sex"] = df["sex"].map(sex_map)
df["survived"] = df["survived"].map(survived_map)
df["pclass"] = df["pclass"].map(pclass_map)
#chuyển sang category
df["embarked"] = df["embarked"].astype("category")
df["sex"] = df["sex"].astype("category")
df["survived"] = df["survived"].astype("category")
#pclass -> cần order lại thứ tự
pclass_order = ['1st', '2nd', '3rd']
df["pclass"] = pd.Categorical(
    df["pclass"],
    categories=pclass_order,
    ordered=True
)
print('\n')
print(df.dtypes)
print('\n')
df.head()
#check pclass order
print(f"\nMin pclass: {df['pclass'].min()}")
print(f"Max pclass: {df['pclass'].max()}")
```
