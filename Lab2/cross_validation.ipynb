```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score, KFold, RepeatedStratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_validate,GridSearchCV
```

```python
#ex 4.1
from IPython.lib.security import random
from sklearn import pipeline
titanic = '/content/drive/MyDrive/DAP/titanic.csv'
df = pd.read_csv(titanic)
features = ['pclass', 'age', 'sibsp', 'parch', 'fare','sex']
target = 'survived'
x = df[features].copy()
y=df[target]
#thiết lập pipline
pipeline = Pipeline([
    ('imputer',SimpleImputer(strategy='median')),
    ('scaler',StandardScaler()),
    ('model',LogisticRegression(max_iter=1000,random_state=42))
])
#Cross validation
#cách 1: 5-Fold
kf = KFold(n_splits=5,shuffle=True,random_state=42)
scores_kf = cross_val_score(pipeline,x,y,cv=kf,scoring='accuracy')
#cách 2 : repeated stratified 5-fold
rskf = RepeatedStratifiedKFold(n_splits=5,n_repeats=10,random_state=42)
scores_rskf = cross_val_score(pipeline,x,y,cv=rskf,scoring='accuracy')
# kiểm tra model
print(f"scores_rskf:{scores_rskf}")
print(f"scores_kf:{scores_kf}")
print('\n')
print(f"Mean Accuracy kf: {np.mean(scores_kf):.4f}")
print(f"Std Dev kf: {np.std(scores_kf):.4f}")
print('\n')
print(f"Mean Accuracy rskf: {np.mean(scores_rskf):.4f}")
print(f"Std Dev rskf: {np.std(scores_rskf):.4f}")
```

```python
#kiểm tra overfitting
cv_results = cross_validate(pipeline,x,y,cv=rskf,scoring='accuracy',return_train_score=True)
mean_train_score = np.mean(cv_results['train_score'])
mean_test_score = np.mean(cv_results['test_score'])
print(f"Mean Train Score: {mean_train_score:.4f}")
print(f"Mean Test Score (CV): {mean_test_score:.4f}")
print(f"Chênh lệch (Train - Test): {mean_train_score - mean_test_score:.4f}")
print('\n')
if mean_train_score - mean_test_score > 0.1: # Nếu train hơn test 10%
    print("Mô hình bị Overfitting nặng.")
elif mean_train_score - mean_test_score > 0.05: # Nếu train hơn test 5%
    print("Có dấu hiệu Overfitting nhẹ.")
else:
    print(" Mô hình ổn định, không bị Overfitting")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import cross_val_score, cross_validate, RepeatedKFold, RepeatedStratifiedKFold, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
# Exercise 4.2
wine = '/content/drive/MyDrive/DAP/wines.csv' ###
df = pd.read_csv(wine) ##
df.columns = df.columns.str.replace(' ','_')
#sửa lỗi type str
if 'type' in df.columns:
  df['type'] = df['type'].map({'red': 0, 'white':1})

x = df.drop('quality',axis=1)
y=df['quality']
#pipline
pipeline = Pipeline([
    ('scaler',StandardScaler()),
    ('model',LogisticRegression(max_iter=1000,random_state=42))
])
#cross validation
#Cách 1: Repeated K-Fold
rkf = RepeatedKFold(n_splits=5,n_repeats=10,random_state=42)
scores_rkf = cross_val_score(pipeline,x,y,cv=rkf,scoring='accuracy')
#cách 2 : Repeated Stratified K-Fold
rskf = RepeatedStratifiedKFold(n_splits=5,n_repeats=10,random_state=42)
scores_rskf = cross_val_score(pipeline,x,y,cv=rskf,scoring='accuracy')
#kiểm tra model
print(f"Repeated K-Fold (No Stratify) - Std Dev: {np.std(scores_rkf):.4f} | Mean: {np.mean(scores_rkf):.4f}")
print(f"Repeated Stratified K-Fold    - Std Dev: {np.std(scores_rskf):.4f} | Mean: {np.mean(scores_rskf):.4f}")
#kiểm tra overfitting
cv_results = cross_validate(pipeline,x,y,cv=rskf,scoring='accuracy',return_train_score=True)
mean_train = np.mean(cv_results['train_score'])
mean_test = np.mean(cv_results['test_score'])
print(f"Mean Train Score: {mean_train:.4f}")
print(f"Mean Test Score:  {mean_test:.4f}")
print(f"Chênh lệch:       {mean_train - mean_test:.4f}")

if mean_train - mean_test > 0.1:
    print("Overfitting nặng!")
else:
    print("Mô hình không bị Overfitting quá mức.")


```
